```python
# plotting the relation between the categorical features and Credit Score
categorical_columns = credit_df.select_dtypes('object').columns.to_list()
categorical_columns.remove("Credit_Score")

for column in categorical_columns:
    plt.figure(figsize=(16, 6))
    sns.countplot(x=column, hue="Credit_Score", data=credit_df)
    plt.title(f"Relation of {column} and Credit Score")
    plt.show()
```
