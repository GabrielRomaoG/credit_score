```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from credit_score_1.cs1_preprocessing import Cs1DataSetPreProcessing
```

```python
credit_df = pd.read_csv("credit_score_dataset_1.csv")

credit_df = Cs1DataSetPreProcessing.process(credit_df)

credit_df.head()
```

```python
credit_df.education.unique()
```

```python
credit_df.describe()
```

```python
credit_df["Credit Score"].unique()
```

```python
# plotting the relation between the categorical features and Credit Score
categorical_columns = credit_df.select_dtypes('object').columns.to_list()
categorical_columns.remove("Credit Score")

for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, hue="Credit Score", data=credit_df)
    plt.title(f"Relation of {column} and Credit Score")
    plt.show()
```

```python

numerical_columns = credit_df.select_dtypes('number').columns

for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=credit_df, x="Credit Score", y=column)
    plt.title(f"Relation between {column} and Credit Score")
    plt.show()
```
